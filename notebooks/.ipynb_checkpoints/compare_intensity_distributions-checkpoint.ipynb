## Compare Intensity Distributions: Style vs Faket Tomograms

Compares pixel intensity distributions between a style tomogram and
style-transferred faket tomograms across multiple SNR values.

For each `train_dir`, histograms and radially-averaged power spectra are computed per tomogram
and averaged across tomograms within that directory. The SNR label is read from the `snr_list` JSON.

**Metrics:**
- Normalised intensity histogram overlay
- Radially-averaged power spectrum (iPSD)
- Summary statistics table per SNR

```python
import sys
sys.path.insert(0, '..')

import json
import numpy as np
import mrcfile
import matplotlib.pyplot as plt
from pathlib import Path
```

### Set paths

```python
PARENT_DIR         = Path(/projects/extern/nhr/nhr_ni/nim00020/dir.project/sage)
faket_base_dir     = PARENT_DIR / "data/simulation/testing/run7/train_dir_0"    
style_dir          = faket_base_dir / "style_tomograms_0"
n_train_dirs       = 10                                                            
simulation_index   = 0                                                             
BINS               = 256
```

### Helper functions

```python
def load_mrc(path):
    with mrcfile.open(path, permissive=True) as mrc:
        return np.copy(mrc.data).astype(np.float32)

def radial_avg_power_spectrum(arr):
    f  = np.fft.fftn(arr)
    ps = np.abs(np.fft.fftshift(f)) ** 2
    center = np.array(ps.shape) // 2
    z, y, x = np.indices(ps.shape)
    r = np.sqrt(
        (z - center[0])**2 + (y - center[1])**2 + (x - center[2])**2
    ).astype(int)
    r_max = int(min(center))
    rps   = np.array([ps[r == ri].mean() for ri in range(r_max)])
    freqs = np.arange(r_max) / r_max
    return rps, freqs

def collect_faket_paths(train_dir):
    """Collect all *_faket.mrc files from a train_dir, handling flat and CZII structures."""
    return sorted((train_dir / "faket_tomograms").rglob("*_faket.mrc"))

def get_mean_snr(train_dir, simulation_index):
    """Read mean SNR from the snr_list JSON for this train_dir."""
    snr_json = train_dir / "snr_list_dir" / f"snr_list_{simulation_index}.json"
    if not snr_json.exists():
        return None
    with open(snr_json) as f:
        data = json.load(f)
    snrs = [v["snr"] for v in data.values() if "snr" in v]
    return float(np.mean(snrs)) if snrs else None
```

### Step 1: Determine global intensity range across all tomograms
Needed so histograms share consistent bins and can be averaged.

```python
print("Computing global intensity range...")

all_p1, all_p99 = [], []

# Style tomogram
style_data = load_mrc(style_path)
all_p1.append(np.percentile(style_data, 1))
all_p99.append(np.percentile(style_data, 99))

# Faket tomograms across all train_dirs
for i in range(n_train_dirs):
    train_dir = base_dir / f"train_dir_{i}"
    for p in collect_faket_paths(train_dir):
        d = load_mrc(p)
        all_p1.append(np.percentile(d, 1))
        all_p99.append(np.percentile(d, 99))

vmin, vmax = min(all_p1), max(all_p99)
print(f"Global intensity range (p1–p99): [{vmin:.4f}, {vmax:.4f}]")
```

### Step 2: Compute distributions per train_dir and average

```python
print("Computing histograms and power spectra...\n")

# --- Style tomogram ---
style_hist, bin_edges = np.histogram(style_data.ravel(), bins=BINS, range=(vmin, vmax), density=True)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
print("Computing style tomogram power spectrum...")
style_rps, style_freqs = radial_avg_power_spectrum(style_data)

# --- Faket tomograms per train_dir ---
results = []  # list of dicts: {snr, mean_hist, mean_rps, freqs, stats}

for i in range(n_train_dirs):
    train_dir = base_dir / f"train_dir_{i}"
    paths     = collect_faket_paths(train_dir)
    snr       = get_mean_snr(train_dir, simulation_index)
    label     = f"SNR={snr:.3f}" if snr is not None else f"train_dir_{i}"

    if not paths:
        print(f"train_dir_{i}: no faket tomograms found, skipping.")
        continue

    print(f"train_dir_{i} ({label}): {len(paths)} tomograms")

    hists, rpss, all_means, all_stds = [], [], [], []
    for p in paths:
        d = load_mrc(p)
        h, _ = np.histogram(d.ravel(), bins=BINS, range=(vmin, vmax), density=True)
        rps, freqs = radial_avg_power_spectrum(d)
        hists.append(h)
        rpss.append(rps)
        all_means.append(d.mean())
        all_stds.append(d.std())

    # Align PSD lengths (trim to shortest)
    min_len = min(len(r) for r in rpss)
    rpss    = [r[:min_len] for r in rpss]
    freqs   = freqs[:min_len]

    results.append({
        "label":     label,
        "snr":       snr,
        "mean_hist": np.mean(hists, axis=0),
        "mean_rps":  np.mean(rpss, axis=0),
        "freqs":     freqs,
        "mean_mu":   np.mean(all_means),
        "mean_std":  np.mean(all_stds),
        "n":         len(paths),
    })

print("\nDone.")
```

### Step 3: Summary statistics table

```python
print(f"{'Label':<20} {'Mean':>10} {'Std':>10} {'N tomos':>8}")
print("-" * 52)
print(f"{'style':<20} {style_data.mean():>10.4f} {style_data.std():>10.4f} {'1':>8}")
for r in results:
    print(f"{r['label']:<20} {r['mean_mu']:>10.4f} {r['mean_std']:>10.4f} {r['n']:>8}")
```

### Step 4: Plot

```python
cmap   = plt.get_cmap("viridis", len(results))
colors = [cmap(i) for i in range(len(results))]

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# --- Histogram ---
ax = axes[0]
ax.plot(bin_centers, style_hist, color="black", linewidth=2, label="style", zorder=10)
for r, c in zip(results, colors):
    ax.plot(bin_centers, r["mean_hist"], color=c, linewidth=1.5, label=r["label"], alpha=0.85)
ax.set_xlabel("Intensity")
ax.set_ylabel("Density")
ax.set_title("Intensity Distribution (averaged per SNR)")
ax.legend(fontsize=8)

# --- Power spectrum ---
ax = axes[1]
min_style_len = len(style_rps)
ax.semilogy(style_freqs[:min_style_len], style_rps[:min_style_len],
            color="black", linewidth=2, label="style", zorder=10)
for r, c in zip(results, colors):
    min_len = min(len(r["freqs"]), min_style_len)
    ax.semilogy(r["freqs"][:min_len], r["mean_rps"][:min_len],
                color=c, linewidth=1.5, label=r["label"], alpha=0.85)
ax.set_xlabel("Spatial frequency (cycles/voxel)")
ax.set_ylabel("Power (log scale)")
ax.set_title("Radially-Averaged Power Spectrum (averaged per SNR)")
ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```
